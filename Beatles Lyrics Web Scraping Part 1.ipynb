```python
#<strong>The Beatles Lyrics</strong>
# Isolate the div that has all the song links
songs_section = soup.find('div', {'id': 'listAlbum'})

# Create an empty list
songs = []

# Loop through each link in the song div
for song_link in songs_section.find_all('a', {'target': '_blank'}):
    # Start with an empty dictionary
    song = {}
    
    # Add name
    song['name'] = song_link.text
    
    # Add slug from the href
    song['slug'] = song_link.get_attribute_list('href')[0].strip('..')
    
    # Add song to our list of songs
    songs.append(song)
songs
```

```python

albums=[]

all_songs=[]

all_lyrics=[]

for song in songs:
    song_slug=song['slug']
    print("https://www.azlyrics.com{}".format(song_slug))
    response = requests.get("https://www.azlyrics.com{}".format(song_slug))
    soup_1 = bs(response.content, 'lxml')
    print(soup_1)
    album=soup_1.find('div', {'class': 'panel songlist-panel noprint'}).b.text
    albums.append(album)
    song=soup_1.find('div', {'class':'col-xs-12 col-lg-8 text-center'}).findChildren('b')[1].text
    all_songs.append(song)
    lyrics=soup_1.find_all('div', {'class': ''})
    all_lyrics.append(lyrics)
```

```python
#Since we forgot to place a sleep on the scraping process, the process terminated before completion. Nevertheless,
#it picked 159 songs.
df = pd.DataFrame({'album': albums, 'song_title': all_songs, 'lyrics': all_lyrics})
```
