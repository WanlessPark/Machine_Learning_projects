```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import time
```

```python
#The same model was used to web scrape all selected Rolling Stones albums and also the last two original Beatles albums:
#"Abbey Road" and "Let It Be". We just need to replace the URL below for each album being scraped.
url= "https://genius.com/albums/The-rolling-stones/Get-yer-ya-ya-s-out"
```

```python
result = requests.get(url)
result
```

```python
soup = bs(result.content, 'lxml')
```

```python
print (soup.prettify)
```

```python
soup.title
```

```python
song_links_section = soup.find_all('div', {'class': 'chart_row-content'})
song_links_section
```

```python
soup.find('a', {'class':"u-display_block"})['href']
```

```python
# Create an empty list
songs_attributes = []

# Loop through each link in the song div
for song_attributes_link in soup.find_all('div', {'class': 'chart_row-content'}):
    # Start with an empty dictionary
    try:
    
        song_attributes = {}
    
    # Add name
        song_attributes['title_2'] = song_attributes_link.text.replace('\n','')
    
    # Add slug
        song_attributes['lyrics_link'] = song_attributes_link.find('a', {'class': 'u-display_block'})['href']
    
    # Add restaurant to our list of restaurants
        songs_attributes.append(song_attributes)
    except:
        print ('error')
        pass
        time.sleep(5)
```

```python
song_attributes
```

```python
songs_attributes
```

```python


all_songs=[]

all_lyrics=[]

for song_attributes in songs_attributes:
    song_slug=song_attributes['lyrics_link']
    print(song_slug)
    response = requests.get(song_slug)
    soup_3 = bs(response.content, 'lxml')
    print(soup_3)
    #album=soup_3.find('div', {'class': 'panel songlist-panel noprint'}).b.text
    #albums.append(album)
    song_attributes=soup_3.find('div', {'class': 'song_body-lyrics'}).text.split('\n\n\n\n')[0]
    all_songs.append(song_attributes)
    lyrics=soup_3.find('div', {'class': 'song_body-lyrics'}).text.split('\n\n\n\n')[1]
    all_lyrics.append(lyrics)
    time.sleep(5)
```

```python
all_lyrics
```

```python
df= pd.DataFrame({'song_title': all_songs, 'lyrics': all_lyrics})
```

```python
df.to_csv('Rolling Stones_lyrics_Decembers_Get_Yer_Ya_Yas_Out.csv', index=False)
```
