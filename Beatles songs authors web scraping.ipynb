```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import time
```

```python
#This site was used to scrape the author of Beatles songs.
url= "https://en.wikipedia.org/wiki/List_of_songs_recorded_by_the_Beatles"

```

```python
result = requests.get(url)
result
```

```python
soup = bs(result.content, 'lxml')
```

```python
print (soup.prettify)
```

```python
soup.title
```

```python
song_link = soup.find_all('th', {'scope': "row"})[3].text
song_link
```

```python
table_1=soup.find_all('table', attrs={'class':"wikitable"})
table = table_1[1]
table
```

```python
table_body = table.find('tbody')
table_body
```

```python
type(table_body)
```

```python
rows = table_body.find('th',{'scope': 'row'})
song_title=rows.text.strip('\n')
song_author=rows.find_next_sibling().find_next_sibling().text.strip('\n')
```

```python
song_title
```

```python


# Create an empty list
songs_attributes = []

# Loop through each link in the song div
for song_attributes_link in table_body.find_all('th', {'scope': 'row'}):
    # Start with an empty dictionary
    song_attributes = {}
    
    # Add name
    song_attributes['title_1'] = song_attributes_link.text.strip('\n')
    
    # Add slug from the href
    song_attributes['author'] = song_attributes_link.find_next_sibling().find_next_sibling().text.strip('\n')
    
    # Add restaurant to our list of restaurants
    songs_attributes.append(song_attributes)
    time.sleep(5)
```

```python
songs_attributes
```

```python
for song in songs_attributes:
    print (song['title_1'])
```
